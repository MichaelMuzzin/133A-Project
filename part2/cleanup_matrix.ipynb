```python
df = pd.read_csv('../survey_results_public_2024.csv' , index_col='ResponseId')
pd.set_option('display.max_columns', 35)
pd.set_option('display.max_rows', 100)
df = df.drop(DROP_FEATURES, axis=1)
country_grp = df.groupby(['Country'])
America = country_grp.get_group('United States of America').copy()


```

```python

def map_employment(cell):
    # Return 0 for missing values
    if pd.isna(cell):
        return 0
    # Split the cell string by comma and remove extra spaces
    options = [opt.strip() for opt in cell.split(';')]
    # Aggregate the mapped values (in this case, summing them)
    return sum(MAPPING_EMPLOYEMENT.get(opt, 0) for opt in options)
America['Employment'] = America['Employment'].apply(map_employment)
```
